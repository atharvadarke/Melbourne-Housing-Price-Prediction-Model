```python
y = new_melb_data.Price
y
```

```python
melb_features = ['Postcode','Landsize','Rooms','Bathroom','Bedroom2','YearBuilt']
X = new_melb_data[melb_features]
```

```python
# Training and Validation sets 
train_X,val_X,train_y,val_y = train_test_split(X,y,random_state=0)
```

```python
melb_model = DecisionTreeRegressor(random_state=0)
melb_fit = melb_model.fit(train_X,train_y)
pred_vals = melb_model.predict(val_X)
melb_fit
```

```python
for x in pred_vals:
    print(x)
```

```python
def get_mae(max_leaf_nodes,train_X,val_X,train_y,val_y):
    global melb_model
    melb_model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes,random_state=1)
    melb_model.fit(train_X,train_y)
    pred_vals = melb_model.predict(val_X)
    mae = mean_absolute_error(val_y,pred_vals)
    return mae
```

```python
max_leaf_nodes_size = [5,10,20,50,100,500,10000]
all_mae = {leaf_nodes:get_mae(leaf_nodes,train_X,val_X,train_y,val_y) for leaf_nodes in max_leaf_nodes_size}
for i in all_mae:
    print(f"Number of leaf nodes:{i} \t\t mean absolute error:{all_mae[i]}")
```

```python
optimal = min(all_mae,key=all_mae.get)
print("Most optimal no. of leaf nodes: %d"%(optimal))
```
